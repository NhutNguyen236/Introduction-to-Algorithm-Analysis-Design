# Selection sort algorithm

```
def sele_sort(A):
  #Sorts a given array by selection sort
  #Input: An array A[0..n − 1] of orderable elements
  #Output: Array A[0..n − 1] sorted in nondecreasing order

  # Get A length
  n = len(A)

  for i in range(0,n-1): #n-2 for last sort
    min = i # i = 0 is the min as default
    for j in range(i+1,n):
      if A[j] < A[min]: #(*)
        min = j
      #print('i=', i, 'j=', j, 'A[min]=', A[min], A)
    A[i],A[min] = A[min],A[i]
  return A

#Implementation|
A = auto_gen()
print('Input array:', '\n', A)
print('----------------------')
print(sele_sort(A))
'''
Time complexity: 
  - 2 for loops, most costly is (*)
  - O(n) = Sum from 0 to n(Sum from i+1 to n) = Sum from 0 to n(n-i) = n^2/2 is in Theta(n^2) => O(n) = Theta(n^2)
'''
```

# Bubble sort algorithm

```
def bubble_sort(A):
  #Sorts a given array by bubble sort
  #Input: An array A[0..n − 1] of orderable elements
  #Output: Array A[0..n − 1] sorted in nondecreasing order

  # Get A length
  n = len(A)
  for i in range(0,n-1):
    for j in range(0,n-1-i):
      if A[j+1] < A[j]: #(*)
        A[j], A[j+1] = A[j+1], A[j]
      #print('i=', i, 'j=', j, A)
  return A

#Implementation
A = auto_gen()
print('Input array:', '\n', A)
print(bubble_sort(A))

'''
Time complexity: 
  - 2 for loops, most costly is (*)
  - O(n) = Theta(n^2)
'''
```

# Sequential search algorithm

```
def seq_search(A,k):
  #Implements sequential search with a search key as a sentinel
  #Input: An array A of n elements and a search key K
  #Output: The index of the first element in A[0..n − 1] whose value is equal to K or −1 if no such element is found

  # Get A length
  n = len(A)

  last = A[n-1]

  A[n-1] = k
  i = 0

  while(A[i] != k): 
    i += 1
  
  # Put the last element back to its own place
  A[n-1] = last
  
  if((i < n-1) or (k == A[n-1])): 
    return i
  else:
    return -1 

#Implementation
A = auto_gen()
print(A)
print(seq_search(A,A[len(A)-1]))

'''
Time complexity: 
  - 1 while loop, most costly is (*)
  - O(n) = Theta(n)
'''
```

# Still Sequential search but with list as input not array :) 

```
def seq_search_list(A,k):
  #Implements sequential search with a search key as a sentinel
  #Input: An array A of n elements and a search key K
  #Output: The index of the first element in A[0..n − 1] whose value is equal to K or −1 if no such element is found

  # Get A length
  n = len(A) - 1 

  A.append(k)
  i = 0

  while(A[i] != k): 
    i += 1
  
  if i < n - 1 : 
    return i
  else:
    return -1

#Implementation
A = list_gen()
print(A)
print(seq_search_list(A,2))
```
